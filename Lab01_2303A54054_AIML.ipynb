```python
people = ["Rithu","Nav","Varsha","Sreeja","Sidhu","Tharun","Harini"]
age = [18, 19, 20, 18, 19, 15,21]
weight = [40, 50, 45, 42, 52, 60, 49]
height = [160, 140, 150, 100, 120, 160,180]
```

```python
plt.scatter(people,age,color="red",marker="*")
plt.title("Name & Age")
plt.xlabel("Names of people")
plt.ylabel("Ages of people")
```

```python
plt.bar(people,age,color="Yellow",width=0.6,edgecolor='k',alpha = 0.6)
plt.title("People's names and their ages",fontsize=20,fontstyle='italic')
plt.xlabel("Names of people",fontsize=16)
plt.ylabel("Ages of people",fontsize=16)
plt.show()
```

```python
plt.hist(weight,color="blue",edgecolor="k",alpha=0.75,bins=5)
plt.title("Histogram of weights & heights of people",fontsize=18)
plt.xlabel("Weights of people")
plt.show()
```
